Removing Columns which are not relevant (Reference purpose)

```python
# EnrolData.to_csv('EnrolData.csv')

# removing data which are not relevant these data are just for referencing purpose hence should be removed
noncol = ['Academic Period','Unique ID','Application Date','Enrolled']
EnrolData = EnrolData.drop(noncol,axis=1)
```

# Data Distribution

```python
plt.figure(figsize=(21,7), dpi=300)
sns.countplot(x='State Province',data=EnrolData)
plt.savefig('State Province.png')
plt.show()
```

```python
# Since the data is categorical and it is just 1.2 % missing data we can impute it with Mode
EnrolData['State Province'].fillna(EnrolData['State Province'].mode()[0],inplace=True)

# EnrolData['State Province'].mode()[0]
```

```python
sns.countplot(x='Admissions Population Description', data = EnrolData)
plt.show()
# Since the data is categorical and it is just 3.2 % missing data we can impute it with Mode
EnrolData['Admissions Population Description'].fillna(EnrolData['Admissions Population Description'].mode()[0],inplace=True)
EnrolData.head(38018)
```

```python
# Since 8.7% of High School GPA data is missing therefore imputing the data with Mode because the data is negatively skewed also GPA has maximum limit of 5 so we cant use MLE method
high=EnrolData['High School GPA']
high=high.dropna()
sns.set_style('darkgrid')
sns.distplot(high,kde=False)
plt.show()
EnrolData['High School GPA'].fillna(EnrolData['High School GPA'].mode()[0],inplace=True)
```

```python
# For institutional aid offered we are assuing zero aid offer for missing value
# Aid=EnrolData['Institutional Aid Offered']
# Aid=Aid.dropna()
# sns.set_style('darkgrid')
# sns.distplot(Aid)
# plt.show()
EnrolData['Sat Verbal'].fillna(EnrolData['Sat Verbal'].mean(),inplace=True)
EnrolData['Sat Total Score'].fillna(EnrolData['Sat Total Score'].mean(),inplace=True)
EnrolData['Sat Mathematics'].fillna(EnrolData['Sat Mathematics'].mean(),inplace=True)
EnrolData['Institutional Aid Offered'].fillna(0,inplace=True)
EnrolData= EnrolData.dropna(axis=0)
```

```python
# box plot for looking outliers we cant remove outliers as there is no error
fig, axs = plt.subplots(ncols=5)
sns.set_style('darkgrid')
sns.boxenplot(x='Admitted',y='High School GPA',data=EnrolData,ax=axs[0])
sns.boxenplot(x='Admitted',y='Sat Verbal',data=EnrolData,ax=axs[1])
sns.boxenplot(x='Admitted',y='Sat Mathematics',data=EnrolData,ax=axs[2])
sns.boxenplot(x='Admitted',y='Sat Total Score',data=EnrolData,ax=axs[3])
sns.boxenplot(x='Admitted',y='Institutional Aid Offered',data=EnrolData,ax=axs[4])
# fig.tight_layout()
plt.show()
```

```python
# looking at unique values of categorical variables
print(EnrolData['State Province'].unique())
print(EnrolData['Student Population'].unique())
print(EnrolData['Admissions Population Description'].unique())
print(EnrolData['Residency Description'].unique())
print(EnrolData['Admissions Athlete'].unique())
```

```python
Aid=EnrolData['Pre-Veterinarian']
Aid=Aid.dropna()
sns.set_style('darkgrid')
sns.distplot(Aid)
plt.show()
```

```python
# Pre-Veterinarian has only one value 0 so dropping the value
EnrolData = EnrolData.drop('Pre-Veterinarian',axis=1)
```

Plot Histogram for all features in one go

```python
# Applying labelEncoder in categorical data
from sklearn.preprocessing import LabelEncoder
laben = LabelEncoder()
EnrolData['State Province'] = laben.fit_transform(EnrolData['State Province'])
EnrolData['Student Population'] = laben.fit_transform(EnrolData['Student Population'])
EnrolData['Admissions Population Description'] = laben.fit_transform(EnrolData['Admissions Population Description'])
EnrolData['Residency Description'] = laben.fit_transform(EnrolData['Residency Description'])
EnrolData['College Description'] = laben.fit_transform(EnrolData['College Description'])
EnrolData['Major Description'] = laben.fit_transform(EnrolData['Major Description'])
EnrolData['Gender'] = laben.fit_transform(EnrolData['Gender'])

EnrolData['Admitted'].replace('Admitted',1,inplace = True)
EnrolData['Admitted'].replace('Denied',0,inplace = True)

print(EnrolData['Admitted'].unique())
```

```python
# Applying Decision Tree and Random forest method
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix, accuracy_score

X_value = EnrolData.iloc[:,:-1].values
Y_Value = EnrolData.iloc [:,-1].values

x_train, x_test, y_train, y_test = train_test_split(X_value, Y_Value, test_size=0.2, random_state=0)


#Using Gini

Binarytree = DecisionTreeClassifier(criterion="gini", splitter="best", max_depth=3)
Binarytree = Binarytree.fit(x_train,y_train)
# print(" Binary Tree score with Gini",Binarytree.score(x_test,y_test))
y_predict = Binarytree.predict(x_test)
print("Accuracy Binary Tree with Gini:",accuracy_score(y_test, y_predict))
print("Confusion matrix \n",confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
fpr, tpr, thresholds= metrics.roc_curve(y_test,y_predict)
auc = metrics.roc_auc_score(y_test,y_predict, average='macro', sample_weight=None)
sns.set_style('darkgrid')
sns.lineplot(fpr,tpr,color ='blue')
plt.show()
```

```python
# Using information gain

Binarytree = DecisionTreeClassifier(criterion="entropy", splitter="best", max_depth=3)
Binarytree = Binarytree.fit(x_train,y_train)
# print(" Binary Tree score with Gini",Binarytree.score(x_test,y_test))
y_predict = Binarytree.predict(x_test)
print("Accuracy Binary Tree with information gain:",accuracy_score(y_test, y_predict))
print("Confusion matrix \n",confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
fpr, tpr, thresholds= metrics.roc_curve(y_test,y_predict)
auc = metrics.roc_auc_score(y_test,y_predict, average='macro', sample_weight=None)
sns.set_style('darkgrid')
sns.lineplot(fpr,tpr,color ='blue')
plt.show()
```

```python
# using random Forest
Forestclassifier = RandomForestClassifier(n_estimators= 80, criterion="entropy", bootstrap = True, max_depth=3, max_features = 'sqrt')
Forestclassifier.fit(x_train,y_train)
y_predict = Binarytree.predict(x_test)
print("Accuracy Random Forest:",accuracy_score(y_test, y_predict))
print("Confusion matrix \n",confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
fpr, tpr, thresholds= metrics.roc_curve(y_test,y_predict)
auc = metrics.roc_auc_score(y_test,y_predict, average='macro', sample_weight=None)
sns.set_style('darkgrid')
sns.lineplot(fpr,tpr,color ='blue')
plt.show()
```

# Hyper Parameter Tuning for Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV
# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 20, stop = 300, num = 10)]
# number of features at every split
max_features = ['auto', 'sqrt']
criterion=['entropy','gini']
bootstrap = [True,False]
# max depth
max_depth = [int(x) for x in np.linspace(2, 50, num = 11)]
max_depth.append(None)
# create random grid
random_grid = {
 'n_estimators': n_estimators,
 'max_features': max_features,
 'max_depth': max_depth,
 'criterion': criterion,
 'bootstrap': bootstrap 
 }
# Random search of parameters
rfc_random = RandomizedSearchCV(RandomForestClassifier(), param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the model
rfc_random.fit(x_train, y_train)
# print results
print(rfc_random.best_params_)
```

```python
# using random Forest
Forestclassifier = RandomForestClassifier(n_estimators= 175, criterion="entropy", bootstrap = False, max_depth=21, max_features = 'sqrt')
Forestclassifier.fit(x_train,y_train)
y_predict = Forestclassifier.predict(x_test)
print("Accuracy Random Forest:",accuracy_score(y_test, y_predict))
print("Confusion matrix \n",confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
fpr, tpr, thresholds= metrics.roc_curve(y_test,y_predict)
auc = metrics.roc_auc_score(y_test,y_predict, average='macro', sample_weight=None)
sns.set_style('darkgrid')
sns.lineplot(fpr,tpr,color ='blue')
plt.show()
```

# Hyper Parameter Tuning for Lasso Model

```python
# Lasso Regression alpha parameter x_train, y_train

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

lasso=Lasso(max_iter=10e5,normalize=True)
parameters={'alpha': [1e-15, 1e-10, 1e-7,1e-3, 1e-2,1e-1,1,5,10,20]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error', cv=5)
lasso_regressor.fit(x_train, y_train)
# print(lasso_regressor.best_score_)
print(lasso_regressor.best_params_)
```

```python
# using Lasso
lasso_final =Lasso(alpha=1e-07)
lasso_final.fit(x_train,y_train)
y_predict = lasso_final.predict(x_test)
print("Accuracy Lasso Regression:",accuracy_score(y_test, y_predict.round()))
print("Confusion matrix \n",confusion_matrix(y_test,y_predict.round()))
print(classification_report(y_test,y_predict.round()))
fpr, tpr, thresholds= metrics.roc_curve(y_test,y_predict.round())
auc = metrics.roc_auc_score(y_test,y_predict, average='macro', sample_weight=None)
sns.set_style('darkgrid')
sns.lineplot(fpr,tpr,color ='blue')
plt.show()
```

# Hyper Parameter Tuning for Ridge Model

```python
# Ridge Regression alpha parameter
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import RidgeClassifier

ridge=RidgeClassifier(max_iter=10e5)
parameters={'alpha': [1e-15, 1e-10, 1e-7,1e-3, 1e-2,1e-1,1,5,10,20,30,40,50,70,100,200]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error', cv=5)
ridge_regressor.fit(x_train, y_train)
# print(ridge_regressor.best_score_)
print(ridge_regressor.best_params_)
```

```python
# using Ridge
ridge_final =RidgeClassifier(alpha=100)
ridge_final.fit(x_train, y_train)
y_predict = ridge_final.predict(x_test)
print("Accuracy Ridge Classifier:",accuracy_score(y_test, y_predict.round()))
print("Confusion matrix \n",confusion_matrix(y_test,y_predict.round()))
print(classification_report(y_test,y_predict.round()))
fpr, tpr, thresholds= metrics.roc_curve(y_test,y_predict.round())
auc = metrics.roc_auc_score(y_test,y_predict, average='macro', sample_weight=None)
sns.set_style('darkgrid')
sns.lineplot(fpr,tpr,color ='blue')
plt.show()
```
